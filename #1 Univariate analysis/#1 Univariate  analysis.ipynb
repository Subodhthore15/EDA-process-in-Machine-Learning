#  Checking normality
 https://towardsdatascience.com/normality-tests-in-python-31e04aa4f411


```python
# by using diffrent graph 
import seaborn as sns
sns.boxplot(df['age'])

# from below plot we can see median lie at centre so age is slightly
#normally distributed or +ve skewed  
```

```python
sns.distplot(df['age'])
```

```python
import seaborn as sns
sns.boxplot(df['bmi'])

# bmi is normally dostributed median  line is at centre but have some outliers  
```

```python
sns.distplot(df['bmi'])
```

```python
import seaborn as sns
sns.boxplot(df['children'])
```

```python
sns.distplot(df['children'])
```

```python
import seaborn as sns
sns.boxplot(df['charges'])

# here median line is below centre so this is  right skewed data(+ve)
```

```python
sns.distplot(df['charges'])
```

```python
from scipy.stats import chisquare
_,p_value=chisquare(df['age'])
print(p_value)

# here p_value is <= 0.05 so reject  H0 so our age attribute is normally distibute 
```

# finding unique value

#  Checking null values


```python
print('Total null value of sex = ',df['sex'].isnull().sum())
print('Total null value of bmi = ',df['bmi'].isnull().sum())
print('Total null value of children = ',df['children'].isnull().sum())
print('Total null value of charges = ',df['charges'].isnull().sum())
print('Total null value of age = ',df['age'].isnull().sum())
```
